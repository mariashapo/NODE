```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labels = pd.read_csv('hydropeaking_noce_labels.csv')
```

```python
labels.columns
```

```python
data = pd.read_csv('hydropeaking_noce.csv', names = labels.columns)
data
```

```python
df = pd.DataFrame(data[:2000])

# Plotting
plt.figure(figsize=(20, 10))

# Loop through the columns, excluding precipitation and temperature for now
for column in df.columns.drop(['t0408 precipitation', 't0408 temperature']):
    print(column)
    plt.plot(df[column], label=column)


plt.plot(df['t0408 precipitation'], label='Precipitation', linestyle='--')
plt.plot(df['t0408 temperature'], label='Temperature', linestyle='--')

plt.title('Hydrological Data Overview')
plt.xlabel('Time (Index)')
plt.ylabel('Measurements')
plt.legend(loc='upper left')
plt.grid(True)

plt.show()
```
