```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(
    'household_power_consumption.txt',
    sep=';'
)

data.insert(0, 'Datetime', pd.to_datetime(data['Date'] + ' ' + data['Time'],  dayfirst=True))
data.drop(['Date', 'Time'], axis=1, inplace=True)

variables = ['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
for col in variables:
    data[col] = pd.to_numeric(data[col], errors = 'coerce')
```

```python
data_sample = data[:2000]
```

```python
plt.figure(figsize=(20, 8))
#sns.lineplot(x='Datetime', y="Global_active_power", data=data_sample, label="Global_active_power") 
sns.lineplot(x='Datetime', y="Voltage", data=data_sample, label="Sub_metering_1") 
plt.title("Global_active_power")
plt.grid(True)
plt.legend()
```

```python
plt.figure(figsize=(20, 8))
sns.lineplot(x='Datetime', y="Global_active_power", data=data_sample, label="Global_active_power") 
#sns.lineplot(x='Datetime', y="Sub_metering_1", data=data_sample, label="Sub_metering_1") 
plt.title("Global_active_power")
plt.grid(True)
plt.legend()
```
