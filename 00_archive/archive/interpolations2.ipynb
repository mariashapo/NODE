```python
import numpy as np
import matplotlib.pyplot as plt

from interpolation import BarycentricInterpolation
```

```python
poly = BarycentricInterpolation(32, kind='chebyshev2', start=0, stop=10)
```

```python
def cauchy(x):
    """Calculate the Cauchy function."""
    return 1.0 / (1.0 + x**2)
```

```python
n = 20
interpolator = BarycentricInterpolation(n=n, kind='chebyshev2', start=-5, stop=5)

# Obtain y values at the nodes
y_at_nodes = cauchy(interpolator.nodes)

x_plot = np.linspace(-5, 5, 400)

interpolator.fit(y_at_nodes)
x_values = np.linspace(-5, 5, 100)  # Points to evaluate the interpolation
y_interpolated = interpolator.evaluate(x_values)

plt.plot(x_values, y_interpolated, label='Interpolated')
plt.scatter(interpolator.nodes, y_at_nodes, color='red', label='Nodes')
plt.legend()
plt.grid()
plt.show()
```
