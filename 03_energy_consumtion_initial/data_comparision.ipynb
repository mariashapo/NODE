```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pickle

with open('cleanup.pkl', 'rb') as file:
    cleanup_data = pickle.load(file)

with open('processing_function.pkl', 'rb') as file:
    preprocess_data = pickle.load(file)
```

```python
cleanup_data.keys()
```

```python
y_c = cleanup_data['y']
X_c = cleanup_data['X']
t_c = cleanup_data['t']

y_p = preprocess_data['y']
X_p = preprocess_data['X']
t_p = preprocess_data['t']
```

```python
#---------------------------PLOTTING RESULTS--------------------------#
plt.figure(figsize=(15, 7))

plt.plot(t_c, X_c, 'b', label='True Data u(t)')  
plt.plot(t_p, X_p, 'r--', label='Predicted Data u(t)') 

plt.xlabel('Time')
plt.ylabel('u(t)')
plt.legend(loc ="lower right")
plt.grid(True)
plt.show()
```

```python
#---------------------------PLOTTING RESULTS--------------------------#
plt.figure(figsize=(15, 7))

plt.plot(t_c, y_c, 'b', label='True Data u(t)')  
plt.plot(t_p, y_p, 'r--', label='Predicted Data u(t)') 

plt.xlabel('Time')
plt.ylabel('u(t)')
plt.legend(loc ="lower right")
plt.grid(True)
plt.show()
```
